```python
class Aluno:
  def __init__(self, nome, ra, curso):
    self.nome = nome
    self.ra = ra
    self.curso = curso
  def apresentar(self):
    print(f'{self.nome}, {self.ra}, {self.curso}')
```

```python
Aluno1 = Aluno('joaquim', '1452142', ' ADS')

print(Aluno1.nome)
print(Aluno1.ra)
Aluno1.apresentar()
```

```python
#Crie uma classe Carro com os atributos
# marca, modelo e ano. e uma metodo descrição()
#que imprime uma frase apresentando o carro.
#inisara outro metodo chamado freiar()
#quando for chamado imprime " O carro freiou"
```

```python
class Carro:
  def __init__(self, marca, modelo, ano):
    self.marca = marca
    self.modelo = modelo
    self.ano = ano

  def descricao(self):
    print(f'Este é um {self.marca} {self.modelo} do ano {self.ano}.')

  def freiar(self):
    print('O carro freiou')
```

```python

Carro1 = Carro('Ford', 'Focus', 2020)

Carro1.descricao()

Carro1.freiar()
```

```python
class Animal:
  def __init__(self, nome):
    self.nome = nome
    def emitirSom(self):
      print("Som generico de animal")
```

```python
#Criar uma cçasse funcionario
#Criar subclasses gerente e vendedor
#A classe funcionario tem atributo
#nome e salario
# metodo calculaSalario()
#a classe Gerente tem o nome do setor
```

```python
class Funcionario:
  def __init__(self, nome, salario):
    self.nome = nome
    self.salario = salario

  def calculaSalario(self):
    print(self.salario)
```

```python
class Gerente(Funcionario):
  def __init__(self, nome, salario, setor):
    super().__init__(nome, salario)
    self.setor = setor

  def calcula_salario(self):

    print(self.salario * 1.5)
```

```python
gerente = Gerente("João", 5000, "Logistica")
gerente.calcula_salario()
```
